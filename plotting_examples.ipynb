Here are a few simple ways to plot data using mapplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Create some data using trig functions for some interesting plots.

```python
df = pd.DataFrame({ 'x': [i for i in range(0,2000)],  
                    'y_axis1': [2* np.sin(i/100) for i in range(0,2000)],
                    'y_axis2': [1* np.cos(i/100) for i in range(0,2000)],
                    'y_axis3': [.001 * np.tan(i/100) for i in range(0,2000)],
                  })
```

Make line plots into three separate charts.

```python
df.plot(x='x', y='y_axis1', color='brown')
df.plot(x='x', y='y_axis2', color='green')
df.plot(x='x', y='y_axis3', color='green')
```

Create three line plots into one combined chart (using different slices of the data):

```python
# Get all rows, but start at column 1000.
# Note how we have 3 line plots by specifying the sliced dataframe.
# We plot 3 lines in one call to plt.plot().
#
plt.plot(df.iloc[:,1:4][:1000])
plt.show()
```

```python
plt.plot(df['y_axis3'])
```

```python
# In this case, we plot each line using 3 lines.
#
plt.plot('x', 'y_axis1', data=df)
plt.plot('x', 'y_axis2', data=df)
plt.plot('x', 'y_axis3', data=df)
```
