We read historical IBM csv data pulled from [yahoo finance](yahoo.finance.com).

```python
import pandas as pd
df = pd.read_csv("IBM.csv")
print(df.shape)
```

We read only the last 349 rows (around a year worth of data) to focus just on recent stock price.

```python
df_year = df.loc[14500:,['Date','Close','Volume']]
print(df_year.shape)
```

We do simple plots for both closing price and trading volume.

```python
df_year.plot(kind='line',y='Close',color='green')
df_year.plot(kind='line',y='Volume',color='blue')
```

We would like more information on the highest closing price and print date around that time.

```python
print(df_year.loc[df_year['Close'].idxmax()])
```

```python
print(df_year.loc[14620:14630])
```
