## Home Oil Usage Analysis (continued)

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
dr = pd.read_csv("dataset/home-oil.csv")
```

```python
# Keep only the part where we have actual data.
#
df = dr.iloc[:,:22]

# Rename one of the columns to something cleaner.
#
df.rename(columns={'2019-2020.1': '2020-2021'}, inplace=True)
```

# Fixing up the data (removing NaNs)

Two methods:

* Turn the DataFrame into a numpy array, turn the NaNs to 0s and then
  back into a DataFrame.
  
* Just use the Pandas fillna() method

```python
# Turn it into a numpy array; trim off the first column since
# those are month labels.
#
arr2 = np.array(df.iloc[:,1:])
```

```python
# For any nan values, turn them into 0.0
#
arr2[np.isnan(arr2)] = 0.0
```

```python
# Turn the result back into a pd dataframe.
#
dr2 = pd.DataFrame(arr2)
```

```python
# Add the Months column.
#
dr2['Months'] = dr.iloc[:,[0]]
```

```python
dr2
```

```python
# Just make an NaN as 0 directly.
#
df = dr.iloc[:,:22].fillna(0.0)
```

```python
df
```

```python
# Get the sum along axis 0 (i.e., rows), but we will omit column 0
# because those are Month strings.  Turn it into a dataframe.
# Essentially, we are seeing each column summed up to get per-season
# oil usage.
#
per_year = pd.DataFrame(df.iloc[:, 1:].sum(axis = 0))

# When you make a DataFrame like this, the column is just 0 so
# you need to rename it to get a nice legend in the chart.
#
per_year.rename(columns={0:'Gallons Oil Per Season'}, inplace=True)
per_year.plot(kind='bar', y='Gallons Oil Per Season', figsize=(12,6),
                             title='Heating Oil Usage', grid=True)
```

```python
# Get the sum along axis 1 (i.e., columns), but we will omit column 0
# because those are Month strings.  Turn it into a dataframe.
# Essentially, we are seeing each row summed up to get total per-month
# oil usage.
#
per_month = pd.DataFrame(df.iloc[:,1:].sum(axis = 1))
per_month.rename(columns={0:'Total Gallons on Month'}, inplace=True)
per_month['Months'] = df.iloc[:,0]
per_month.plot(kind='bar', x='Months', figsize=(12,6),
                             title='Oil Usage by Month', grid=True)
per_month
```

```python
per_month['Average Gallons'] = per_month['Total Gallons on Month'] / 21
per_month.columns
per_month.plot(kind='bar', x='Months', y='Average Gallons', figsize=(12,6),
                             title='Average Oil Usage by Month', grid=True)
```

```python
# Let's look at a sample of monthly electric bills that go with
# the oil consumption from above.
#
er = pd.read_csv("dataset/electric.csv")
```

```python
er
```

```python
er.columns
```

```python
# We just want the dates and the amount spent on electricity.
#
electric = er[['Date', 'Amount']]

# Convert date string to actual date so we can do queries
# based on time.
#
electric.Date = pd.to_datetime(electric.Date)
electric.Amount = electric.Amount.abs()
electric.Amount.max()
```

```python
year = 2014
for year in range(2009, 2021, 1):
  el = electric [(electric.Date >= f'{year}/01/01') & (electric.Date <= f'{year}/12/31')]
  ave = el.Amount.mean()
  el.plot(kind='bar', x='Date', figsize=(12,6), title=f'{year} Electricity Cost/Month (ave=${ave:.2f})', grid=True, ylim=(100,300))
```
