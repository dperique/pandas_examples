```python
import numpy as np
import pandas as pd
```

```python
# Read somebody's oil consumption data
raw_hh = pd.read_csv("datasets/home-oil.csv")
```

```python
# Note that column 22 and out has no data
raw_hh.iloc[:,22:]
```

```python
# Get a dataframe composed of just real data
hh = raw_hh.iloc[:,:22]
```

```python
# Ensure it contains data
hh
```

```python
# Fixup the months column
hh.rename(columns = {"Unnamed: 0": "Months"}, inplace=True)
hh.rename(columns = {"2019-2020.1": "2020-2021"}, inplace=True)

# Get rid of nans.
hh.replace(np.nan, 0, inplace=True)
```

```python
# Ensure the dataframe looks like how you want it
hh
```

```python
# Sum up a column and return a scalar.
sum(hh['2000-2001'])
```

```python
# Show all columns.
hh.columns
```

```python
# Display two lists: yearRange and Total oil consumption.
# Realize this is not wrong -- but it can be more easily done using
# pandas (as shown later).
col1 = []
col2 = []
for i in hh.columns:
    if "Months" in i:
        continue
    col1.append(i)
    col2.append(sum(hh[i]))
    print(i, f"{sum(hh[i]):9,.2f}")
```

Use pandas to display the same thing as above.

NOTEs:
* We trimmed off the firist row (because it was Months)

```python
hh.sum(axis=0)[1:]
```

```python
# I create a new dataframe using the colums I created above and some
# other random data.
h1 = pd.DataFrame()
h1['yearRange'] = col1
h1['gallons'] = col2
from random import randint
h1['other'] = [randint(0,1000) for i in range(0,len(h1))]
```

```python
h1
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot in two separate graphs
h1.plot(kind='bar',x='yearRange',y='gallons',color='green')
h1.plot(kind='line',x='yearRange',y='other',color='blue')
plt.show()
```

```python
# Multiple lines one graphs
plt.plot('yearRange', 'gallons', data=h1)
plt.plot('yearRange', 'other', data=h1)
plt.legend()
plt.show()
```

```python
# Another way to get total oil consumption per yearRange.
# Note how the Months column was added (as strings)
hh.sum()
```

```python
# Average (mean) consumption per month
hh.mean(axis=1)
```

```python
plt.xticks([i for i in range(12)], hh['Months'])
plt.plot(hh.mean(axis=1), label="gallons")
plt.title("Average Gallons of Oil Per Month")
plt.xlabel("Months")
plt.ylabel("Gallons of Oil")
plt.legend()
plt.show()
```
