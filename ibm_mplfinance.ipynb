```python
%matplotlib inline
import pandas as pd
import mplfinance as mpf
```

```python
file = "dataset/IBM.csv"
data = pd.read_csv(file)
```

```python
data.columns
```

```python
data.Date = pd.to_datetime(data.Date)
```

```python
# Take a subset of the data so it's easier to process
#
df = data[data['Date'] > "2017/04/07"]
```

## Setting the Date as the index

```python
df = df.set_index('Date')
```

## Adding charts with little effort

```python
mpf.plot(df, type='line', volume=True)
```

```python
# Plot just the month of May 2020.
#
# As plain line.
mpf.plot(df.loc["2020-05"], type='line', volume=True, title="IBM in May 20202")

# As line with high/low/close
mpf.plot(df.loc["2020-05"], volume=True)

# As line with high/low/close with candlestick
mpf.plot(df.loc["2020-05"], type='candle', volume=True)
```

```python
# As line with high/low/close with candlestick, with moving average
mpf.plot(df.loc["2020-05":"2020/09"], type='candle', mav=20, volume=True)
```

```python
# As line with high/low/close with candlestick, with moving average.
# Get rid of the front and back gaps (tight_layout=True)
# Set the figure ration to be 20 wide x 12 high (figratio)
# Set the style to yahoo mode (style)
mpf.plot(df.loc["2020-05":"2020/09"], type='line', mav=20, volume=True,
                                      title="IBM May-Sep 2020",
                                      tight_layout=True,
                                      figratio=(20,12),
                                      style='yahoo')
```
